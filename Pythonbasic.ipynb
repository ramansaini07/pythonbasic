2.What is an interpreter in Python?
 - An interpreter in Python is a program that executes Python code line by line. Unlike a compiler, which translates the entire code into machine language before execution, an interpreter processes and runs the code directly, translating it into a format that the computer can understand as it goes.

    This allows for immediate feedback, making Python highly interactive and great for debugging and testing.

3.What are pre-defined keywords in Python?
 - Pre-defined keywords in Python are reserved words that have special meanings and cannot be used as identifiers (such as variable names or function names). These keywords define the syntax and structure of the Python language. For example:-If,int,break,class,etc.

4.Can keywords be used as variable names?
 - No, keywords cannot be used as variable names in Python. Since keywords have special meanings and are part of the language's syntax, they are reserved for specific functions and cannot be redefined or used as identifiers.


5.What is mutability in Python?
 - Mutability in Python refers to whether the value of an object can be changed after it is created.It can be modified after they are created. For example, lists, dictionaries, and sets are mutable. You can change their contents (like adding, removing, or updating elements) without creating a new object.

6. Why are lists mutable, but tuples are immutable?
- Lists are mutable because they are designed to allow modification of their elements, which is useful for cases where the data needs to change over time (e.g., adding or removing items). Tuples, on the other hand, are immutable to ensure data integrity. Once created, they cannot be altered, making them suitable for fixed collections of data, where you want to prevent accidental changes.

7.What is the difference between “==” and “is” operators in Python?
 - == checks if the values of two objects are the same (value comparison). is checks if two variables reference the same object in memory (identity comparison).

8.What are logical operators in Python?
- Logical operators are used to combine conditional statements:

- and: Returns True if both conditions are true. or: Returns True if at least one of the conditions is true. not: Returns the opposite boolean value of the condition.

9. What is type casting in Python?
- Type casting is the process of converting one data type to another. In Python, you can explicitly change data types using functions like int(), float(), str(), etc.
10.What is the difference between implicit and explicit type casting?
-  Implicit type casting (also called coercion) happens automatically by Python when you mix data types in an expression, e.g., adding an integer and a float. Explicit type casting requires you to manually convert the data type using functions like int(), float(), str(), etc.
11.What is the purpose of conditional statements in
Python?
 - Conditional statements allow you to control the flow of execution based on certain conditions. They help in making decisions by executing different blocks of code depending on whether a condition is true or false.

12.How does the elif statement work?
- The elif (else if) statement is used to check multiple conditions in sequence. If the if condition is False, the elif condition(s) are checked, and the first True block is executed.


13.What is the difference between for and while loops?
- for loops are typically used when you know the number
of iterations or when iterating over a sequence (e.g., list, tuple). while loops are used when you want to loop as long as a condition is True, and the number of iterations is not known in advance.

14.Describe a scenario where a while loop is more suitable than a for loop?
- A while loop is more suitable when the number of iterations is not known in advance, and the loop should continue until a specific condition is met. For example, reading data from a file until it’s empty or waiting for user input until it matches a certain condition.

```python
#14. Write a program to reverse a string using a while loop
str = input("Enter a String : ")
str1 = ""
i = len(str)
while i>0 :
  str1 += str[i-1]
  i = i-1
print(str1)
```

```python
# 15.Write a Python program to calculate the factorial of a number provided by the user using a while loop.
user = input("Enter the number to find factorial : ")
user = int(user)
user1 = 1
while user>1:
  user1 = user1*user
  user = user-1
print("factorial of given number  is : ", user1)
```
